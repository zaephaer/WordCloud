# Word Cloud: Terjemahan Surah Al-Mulk (30 ayat)

```python
import numpy as np  # useful for many scientific computing in Python
import pandas as pd # primary data structure library
from PIL import Image # converting images into arrays

%matplotlib inline
import matplotlib as mpl
import matplotlib.pyplot as plt

from wordcloud import WordCloud, STOPWORDS
print ('Wordcloud is installed and imported!')
```

```python
#alice_novel = open('../Word Cloud/alice_novel.txt', 'r').read()  
almulk_tafsir = open('../Word Cloud/al-mulk.txt', 'r').read()  
print ('File found and open')
```

```python
stopwords = set(STOPWORDS)
```

```python
#alice_wc = WordCloud(
almulk_wc = WordCloud(
    background_color='white',
    max_words=10000,
    stopwords=stopwords
)

almulk_wc.generate(almulk_tafsir)

fig = plt.figure()
fig.set_figwidth(14) # set width
fig.set_figheight(18) # set height

plt.imshow(almulk_wc, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
stopwords.add('yang')
stopwords.add('dan')
stopwords.add('kamu')
stopwords.add('mereka')
stopwords.add('tidak')
stopwords.add('itu')
stopwords.add('kami')
stopwords.add('dengan')
stopwords.add('kepada')
stopwords.add('di')
stopwords.add('orang')
stopwords.add('serta')
stopwords.add('ia')
stopwords.add('akan')
stopwords.add('maka')

almulk_wc.generate(almulk_tafsir)

fig = plt.figure()
fig.set_figwidth(14) # set width
fig.set_figheight(18) # set height

plt.imshow(almulk_wc, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
almulk_mask = np.array(Image.open('../Word Cloud/doa.png'))  

almulk_wc = WordCloud(background_color='white', max_words=10000, mask=almulk_mask, stopwords=stopwords)
almulk_wc.generate(almulk_tafsir)

# display the word cloud
fig = plt.figure()
fig.set_figwidth(14) # set width
fig.set_figheight(18) # set height

plt.imshow(almulk_wc, interpolation='bilinear')
plt.axis('off')
plt.show()
```
